```python
from simulation import MichelsonSimulation
import matplotlib.pyplot as plt
import numpy as np
```

## Local interference of single light source

```python
%%time

simulation = MichelsonSimulation()
simulation.initialMirrorG([0, 0, 0], [-1, 1, 0])
simulation.initialMirrorM1([0, 100, 0], [0, -1, 0])
simulation.initialMirrorM2([100-1e-3, 0, 0], [-1, 0, 0])
simulation.insertSource([-20, 0, 0], 600)
print(simulation.getImageSourceList())
simulation.changeToLocal()
simulation.localInterference()

pattern = simulation.localInterference()
pattern = np.array([entry[2] for entry in pattern]).reshape(100,100,-1).sum(axis=2)
plt.imshow(pattern)
```

## Nonlocal interference of two light sources

```python
%%time

simulation = MichelsonSimulation()
simulation.initialMirrorG([0, 0, 0], [-1, 1, 0])
simulation.initialMirrorM1([0, 100, 0], [0, -1, 0])
simulation.initialMirrorM2([100-1, 0, 0], [-1, 0, 0])
simulation.insertSource([-20, 0, 0], 600)
simulation.insertSource([-20, 0, 10], 600)
simulation.changeToNonlocal()

pattern = simulation.nonlocalInterference()
pattern = np.array([entry[2] for entry in pattern]).reshape(100,100,-1).sum(axis=2)
plt.imshow(pattern)
```

## White light local interference

```python
%%time

simulation = MichelsonSimulation()
simulation.initialMirrorG([0, 0, 0], [-1, 1, 0])
simulation.initialMirrorM1([0, 100, 0], [0, -1, 0])
simulation.initialMirrorM2([100, 0, 0], [-1, 1e-6, 0])

for lam in np.arange(300, 700+1, 40):
    simulation.insertSource([-20, 0, 0], lam)

simulation.changeToLocal()

print(simulation.getImageSourceList())

pattern = simulation.localInterference()
pattern = np.array([entry[2] for entry in pattern]).reshape(100,100,-1).sum(axis=2)
plt.imshow(pattern)
```
